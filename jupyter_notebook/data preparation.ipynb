```python
%%sql 
drop table if exists geoannotator.flighttrace_albers;
create table geoannotator.flighttrace_albers
(
    easting_albers text,
    northing_albers text,
    albers text,
    flight text,
    line text,
    fiducial text,
    longitude_gda94 text,
    latitude_gda94 text
)
```

```python
%%sql
insert into geoannotator.flighttrace_albers
with tmp1 as
(
	select 
		ST_AsText (
						st_transform(
							ST_SetSRID(ST_MakePoint(longitude::float,latitude::float), 
							4283),
							3577
									)		
				  ) as albers,	  
	*
	from geoannotator.flighttrace	
)
select 
	regexp_replace(albers,'POINT\(([^ ]+) ([^ ]+)\)','\1'),
	regexp_replace(albers,'POINT\(([^ ]+) ([^ ]+)\)','\2'),
	*
from 
	tmp1
```

```python
#find size of the last tile
%sql select st_width(rast),st_height(rast)  from geoannotator.satellite_tif where rid=104448
```

```python
#generate a downsampled version of satellite tiff
baseimg=%sql select st_aspng(st_union(rast)) from (select st_resample(rast,sample) as rast from geoannotator.satellite_tif as l, (select st_resample(rast,3,3) as sample from geoannotator.satellite_tif where rid=1) as r) tmp
```

```python
#get the size of downsampled satellite image
from io import BytesIO
im = BytesIO(baseimg[0][0])
im = Image.open(im)
im.size
```

```python
#project flight trace onto downsampled satellite image
%%sql 
drop table if exists geoannotator.flighttrace_image;
select distinct 
    ((easting_albers::float-(-38179.126373639))/1390229.999999999*(814-2-1))::int as x,
    ((northing_albers::float-(-2703261.59099128))/1034100.00000000*(1152-30-1))::int as y
into geoannotator.flighttrace_image
from geoannotator.flighttrace_albers
```

```python
#obtain the projected trace
trace=%sql select * from geoannotator.flighttrace_image
```

```python
#display the trace on the satellite image

from io import BytesIO
im = BytesIO(baseimg[0][0])
im = Image.open(im)
pixelvalues = im.load()
for z in trace:
    x = z[0]
    y = z[1]
    color = (255,255,0)
    pixelvalues[x,1152-y]=color
    pixelvalues[x+1,1152-y]=color
    pixelvalues[x,1152-y-1]=color
    pixelvalues[x+1,1152-y-1]=color
im.show()
```
