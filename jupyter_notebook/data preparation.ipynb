```python
%%sql 
drop table if exists geoannotator.flighttrace_albers;
create table geoannotator.flighttrace_albers
(
    easting_albers text,
    northing_albers text,
    albers text,
    flight text,
    line text,
    fiducial text,
    longitude_gda94 text,
    latitude_gda94 text
)
```

```python
%%sql
insert into geoannotator.flighttrace_albers
with tmp1 as
(
	select 
		ST_AsText (
						st_transform(
							ST_SetSRID(ST_MakePoint(longitude::float,latitude::float), 
							4283),
							3577
									)		
				  ) as albers,	  
	*
	from geoannotator.flighttrace	
)
select 
	regexp_replace(albers,'POINT\(([^ ]+) ([^ ]+)\)','\1'),
	regexp_replace(albers,'POINT\(([^ ]+) ([^ ]+)\)','\2'),
	*
from 
	tmp1
```

```python
#generate a downsampled version of satellite tiff
baseimg=%sql select st_aspng(st_union(rast)) from (select st_resample(rast,sample) as rast from geoannotator.satellite_tif as l, (select st_resample(rast,1e3,1e3) as sample from geoannotator.satellite_tif where rid=1) as r) tmp
```

```python
#get the size of downsampled satellite image
from io import BytesIO
from PIL import Image
im = BytesIO(baseimg[0][0])
im = Image.open(im)
im = im.transpose(Image.FLIP_TOP_BOTTOM)
print(im.size)
im.show()
```

```python
%%sql
drop table if exists geoannotator.resample_reference;
select st_resample(rast,1e3,1e3) as rast 
into geoannotator.resample_reference
from 
(select rast, rank() over(order by st_upperleftx(rast), st_upperlefty(rast)) as rank from geoannotator.satellite_tif) as tmp
where rank=1;
```

```python

%%sql 
--project flight trace onto downsampled satellite image
drop table if exists geoannotator.flighttrace_image;
select distinct 
    (ST_WorldToRasterCoord(rast, easting_albers::float, northing_albers::float)).* 
into geoannotator.flighttrace_image
from geoannotator.flighttrace_albers, geoannotator.resample_reference
```

```python
#obtain the projected trace
trace=%sql select * from geoannotator.flighttrace_image
```

```python
#display the trace on the satellite image

from io import BytesIO
im = BytesIO(baseimg[0][0])
im = Image.open(im)
pixelvalues = im.load()
for z in trace:
    x = z[0]
    y = z[1]
    color = (255,255,0)
    pixelvalues[x,y]=color
    pixelvalues[x+1,y]=color
    pixelvalues[x,y+1]=color
    pixelvalues[x+1,y+1]=color

im = im.transpose(Image.FLIP_TOP_BOTTOM)    
im.show()
```
